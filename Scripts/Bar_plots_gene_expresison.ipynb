## Loading data 

This data was obtained from the database BloodSpot : https://pubmed.ncbi.nlm.nih.gov/37933860/ , The data consist of Ing4 gene expression in inmature hematopoietic cells such as LT-HSCs, ST-HSCs, LMPPs, CLPs, etc.

```python
data = pd.read_excel('C:/Users/amade/Documents/UofSC/Bioinformatics/1460728_s_at.xlsx')
data.head(5)
```

## Plotting 

```python
fig, ax = plt.subplots(1,figsize=(8,6))
sns.set(style="whitegrid", color_codes=True)
ax=sns.swarmplot(x="Cell", y="LogFC", data=data, color='black',hue=None,marker="$\circ$", edgecolor="face", s=9)

ax.set_ylabel('Y-axis Label', labelpad=22) 
ax.set_ylabel('Y-axis Label', size=13) 
plt.ylim(5.2, 7.5)
ax.set_xlabel('X-axis Label', size=0) 

# Overlay the average as a shadow using a box plot
sns.boxplot(x="Cell", y="LogFC", data=data, showmeans=True, meanline=True, meanprops={'linestyle': '--', 'color': 'red'})

#plt.xlabel("n")
plt.ylabel("log2 expression")
plt.xticks(fontsize=13.8)
plt.yticks(fontsize=13)
plt.show()
```
