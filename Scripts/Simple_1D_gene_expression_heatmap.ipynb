## Loading Libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import csv 
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.colors import ListedColormap
```

## Importing dataset 

```python
data=pd.read_csv("C:/Users/amade/Desktop/200_most_genes.csv")
data.head()
```

```python
# Obtaining both columns and extracting them into two independent variables
Genes= data.iloc[:, 0]  
LogFC=data.iloc[:,1]
```

```python
# Transformation to array 
LogFC1 = LogFC.to_numpy() #transfroming into an array 
# Reshaping the matrix to 1D array 
change=LogFC1.reshape(-1, 1)
## Transforming list of genes into  a list
list_genes = Genes.values.tolist() 
# Transpossing matrix to plot 
most_expressed = change.T
```

```python
## Plotting most expressed genes 
fig, ax = plt.subplots(figsize=(30,2.3))
im = ax.imshow(most_expressed,origin='lower',aspect='auto',interpolation='nearest',cmap="jet",vmin=1.5, vmax=8)#it hides the y-axis labels
ax.set_xticks([])
ax.set_yticks([])
for i, name in enumerate(list_genes):
    plt.text(i, 0.8, name, ha='center', fontsize=10, color='black',rotation=90, rotation_mode='anchor')
plt.ylabel('LogFC',size=15,fontweight='bold')
divider = make_axes_locatable(ax)
colorbar_axes = divider.append_axes("right",size="3%",pad=0.03)
# Using new axes for colorbar
plt.colorbar(im, cax=colorbar_axes)
```
