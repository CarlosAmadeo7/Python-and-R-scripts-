## Libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Calculating the dinucleotide frequency of a DNA sequence

```python
freq_list=[]
with open("C:/Users/amade/Documents/UofSC/DNA_sequence.txt", 'r') as ins:
    array=[]
    for line in ins:
        array.append(line.replace('\n',''))
    freq_list.append((array[0].count('AA')/len(array[0]))*100)
    freq_list.append((array[0].count('AC')/len(array[0]))*100)  
    freq_list.append((array[0].count('AG')/len(array[0]))*100)  
    freq_list.append((array[0].count('AT')/len(array[0]))*100) 
    freq_list.append((array[0].count('CA')/len(array[0]))*100)
    freq_list.append((array[0].count('CC')/len(array[0]))*100)
    freq_list.append((array[0].count('CG')/len(array[0]))*100)
    freq_list.append((array[0].count('CT')/len(array[0]))*100)
    freq_list.append((array[0].count('GA')/len(array[0]))*100)
    freq_list.append((array[0].count('GC')/len(array[0]))*100)
    freq_list.append((array[0].count('GG')/len(array[0]))*100)
    freq_list.append((array[0].count('GT')/len(array[0]))*100)
    freq_list.append((array[0].count('TA')/len(array[0]))*100)
    freq_list.append((array[0].count('TC')/len(array[0]))*100)
    freq_list.append((array[0].count('TG')/len(array[0]))*100)
    freq_list.append((array[0].count('TT')/len(array[0]))*100)
    
print(freq_list)   # Dinucleotide frequency in percentage 
```

```python
freq_list_1 = [ 5.6, 7.8, 10.555, 4.9555, 5.888, 8.999, 6.18, 5.7, 8.6, 2.94, 6.208,8.554,10.555, 5.88, 6.88, 7.66]
# Dinucleotide frequencies of a second Dna sequence
```

## Calculating the dinucloetide differential frequency of two sequences

```python
# Calculating the dinucleotide differential frequency.
Differential_frequency = [x - y for x in freq_list for y in freq_list_1]

# Reshaping the dinucleotide difference into a 2D array.
differential_array = np.array(Differential_frequency).reshape(len(freq_list), -1)

# Creating a DataFrame.
df = pd.DataFrame(differential_array, columns=[f'Dinucleotide_difference_{i+1}' for i in range(len(freq_list_1))])

# Printing the DataFrame
df
```

## Plotting

```python
#Labels of the axes

xlabs = ["AA", "AC", "AG", "AT",
         "CA", "CC", "CG", "CT","GA", "GC", "GG", "GT","TA", "TC", "TG", "TT"]
ylabs = ["AA", "AC", "AG", "AT",
         "CA", "CC", "CG", "CT","GA", "GC", "GG", "GT","TA", "TC", "TG", "TT"]
```

```python
fig, ax = plt.subplots(figsize=(6,6))
im = ax.imshow(df,cmap='RdBu')

ax.set_xticks(np.arange(len(xlabs)), labels = xlabs)
ax.set_yticks(np.arange(len(ylabs)), labels = ylabs)
ax.tick_params(axis='both', labelsize= 14) 

plt.setp(ax.get_xticklabels(), rotation = 90, ha = "right", rotation_mode = "anchor")

plt.title('Differential Dinucleotide frequency bettween two DNA sequence', size=14,  x=0.55, y=1.05)
#plt.ylabel('First_sequence',size=13,x=-1, y=0.5)
#plt.xlabel('Second_sequence',size=14,x=0.5, y=10)

cbar = ax.figure.colorbar(im, ax = ax,fraction= 0.046, pad=0.04)
plt.show() 
```
